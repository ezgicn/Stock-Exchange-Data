```python
import yfinance as yf
import pandas as pd
```

```python
# Define parameters
ticker = "AAPL"  # Replace with your desired stock ticker
start_date = "2015-01-01"
end_date = "2024-12-31"
output_file = "apple_stock_data.csv"

# Fetch stock data from Yahoo Finance
df = yf.download(ticker, start=start_date, end=end_date)

# Reset the index to ensure 'Date' is a column and save in a clean format
df.reset_index(inplace=True)

# Save the cleaned data to a CSV file
df.to_csv(output_file, index=False)  # Do not save the index

print(f"Cleaned data saved to {output_file}")
```

```python
# Load the CSV file
df = pd.read_csv(output_file)

# Display the first few rows
print(df.head())

# Validate the structure
print(df.info())
```

```python
# Drop the first row containing incorrect headers
df_cleaned = df.iloc[1:].copy()

# Rename columns for consistency
df_cleaned.columns = ["Date", "Close", "High", "Low", "Open", "Volume"]

# Convert 'Date' to datetime and other columns to numeric
df_cleaned["Date"] = pd.to_datetime(df_cleaned["Date"])
numeric_columns = ["Close", "High", "Low", "Open", "Volume"]
df_cleaned[numeric_columns] = df_cleaned[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Display the cleaned data
df_cleaned.head()
```

```python
# Define the output file path
cleaned_file_path = "apple.csv"

# Save the cleaned DataFrame to a CSV file
df_cleaned.to_csv(cleaned_file_path, index=False)

print(f"Cleaned data saved to {cleaned_file_path}")
```
